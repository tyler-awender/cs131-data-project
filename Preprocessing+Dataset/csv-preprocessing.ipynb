```python
import os
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("Spotify_Youtube.csv")

# Drop not needed columns
df.drop(['Description', 'Licensed', 'Url_youtube', 'Url_spotify', 'Uri' ], axis=1, inplace=True)
df.dropna(axis=0,inplace=True)
# drop columns where crucial info is nan
filtered_df = df[df['Likes'].notna() & df['Comments'].notna() & df['Stream'].notna() & df['Title'].notna() & df['Danceability'].notna() & df['Valence'].notna()]

# Get the path to the user's Downloads folder
downloads_folder = os.path.expanduser("~/Downloads")

# Save the CSV file to the user's Downloads folder
output_file_path = os.path.join(downloads_folder, "New_Spotify_Youtube.csv")
filtered_df.to_csv(output_file_path, index=False, sep='|')

# Print the path where the file is saved
print("File saved to:", output_file_path)
```
