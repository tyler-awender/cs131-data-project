```python
import matplotlib.pyplot as plt

# Data
categories = ['Official', 'Unofficial']
likes = [809456, 174047]
comments = [33602, 7447.09]

# Plotting Likes
plt.figure(figsize=(10, 6))
plt.bar(categories, likes, color=['skyblue', 'salmon'])
plt.xlabel('Video Type')
plt.ylabel('Average Likes')
plt.title('Average Likes for Official and Unofficial Videos')

# Adding numbers on top of the bars for Likes
for i in range(len(categories)):
    plt.text(i, likes[i] + 10000, likes[i], ha='center', va='bottom')

plt.show()

# Plotting Comments
plt.figure(figsize=(10, 6))
plt.bar(categories, comments, color=['skyblue', 'salmon'])
plt.xlabel('Video Type')
plt.ylabel('Average Comments')
plt.title('Average Comments for Official and Unofficial Videos')

# Adding numbers on top of the bars for Comments
for i in range(len(categories)):
    plt.text(i, comments[i] + 400, comments[i], ha='center', va='bottom')

plt.show()
```
